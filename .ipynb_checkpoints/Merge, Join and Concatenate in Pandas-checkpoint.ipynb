```python
import pandas as pd
```

```python
df1 = pd.read_csv('LOTR.csv')
df1
```

```python
df2 = pd.read_csv('LOTR 2.csv')
df2
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how = 'outer', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how = 'left', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how = 'right', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how = 'cross')
```

```python

df1.join(df2,on = ['FellowshipID'], lsuffix = '_x',rsuffix = '_y', how = 'left')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),
                                         lsuffix = '_x',rsuffix = '_y', how = 'outer')
df4
```

```python
pd.concat([df1,df2],axis = 0, join = 'outer')
```
