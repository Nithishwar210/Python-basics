```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Ice Cream Ratings.csv')
df = df.set_index('Date')
df
```

```python
df.plot(kind = 'line', subplots = True, title = 'Ice Creame Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df[['Flavor Rating','Texture Rating','Overall Rating'] ].plot.barh(stacked = True)
```

```python
df.plot.scatter(x = 'Overall Rating', y = 'Texture Rating', s = 100, c = 'red')
```

```python
df.plot.hist(bins = 55)
```

```python
df.plot.box()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y = 'Flavor Rating',figsize = (10,10))
```
